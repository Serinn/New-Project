<h3 style="color:#790699;">Removing rows with missing values</h3>

```python
#New dataframe with no missing values by dropping the NAN values
new_df = df.dropna()
new_df
```

<h4 style="color:#111111;">So, we made sure that we have all the features with no missing (NAN) values. </h>


```python
#Change-InvoiceDate- format from object(string) to timeStamp one
new_df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format = '%d-%m-%y %H:%M')
new_df
```

```python
#Calculate Total Price by multiplying units and quantity
new_df['Total_Price']=new_df['Quantity']*new_df['UnitPrice']
new_df.head(10)
```

```python
#To understand the data more clearly, we display the describtive values
new_df.describe()
```

<h4 style="color:#111111;">From the descriptive values, we noticed that: <br>
    <ul>
  <li>Quantity has negative values, which is meaningless actually.</li>
  <li>UnitPrice has zero values... Free products. </li>
</ul> </h>


<h3 style="color:#111111;">In result, we are consedring values that have quantity>0 </h>

```python
#Removing qantities with negative values
new_df = new_df[new_df.Quantity > 0]
```

```python
new_df.describe()
```

```python
#Count the number of unique values in each column
def unique_counts(new_df ):
    for i in new_df.columns:
        count = new_df[i].nunique()
        print(i, ": ", count)
unique_counts(new_df)
```

<h3 style="color:#790699;">Add the columns - Month, Day and Hour for the invoice</h3>

```python
#Splitting the invoiceDate column into 4 separate columns
new_df.insert(loc=5, column='year_month', value=new_df['InvoiceDate'].map(lambda x: 100*x.year + x.month))
new_df.insert(loc=6, column='month', value=new_df.InvoiceDate.dt.month)
# +1 to make Monday=1.....until Sunday=7
new_df.insert(loc=7, column='day', value=(new_df.InvoiceDate.dt.dayofweek)+1)
new_df.insert(loc=8, column='hour', value=new_df.InvoiceDate.dt.hour)
```

<h3 style="color:#790699">Removing Duplicates </h3>

<h3 style="color:#790699">RFM analysis </h3>


```python
import datetime as dt
#Store the present date in present var
present = dt.datetime(2011,12,10)
```

<h4 style="color:#111111;">Recency, frequency and monetary: <br>
    <ul>
  <li>Recency: the number of days between present date and date of last purchase each customer.</li>
  <li>Frequency:  the number of orders for each customer. </li>
   <li>Monetary: the sum of purchase price for each customer.</li>
</ul> </h4>

```python
#Calculate recency, frequency and monetary according to definitions above
rfm= new_df.groupby('CustomerID').agg({'InvoiceDate': lambda date: (present - date.max()).days,
                                        'InvoiceNo': lambda num: len(num),
                                        'Total_Price': lambda price: price.sum()})
```

```python
#Display RFM columns names and datatype 
rfm.columns
```

```python
# Change the name of columns
rfm.columns=['Recency','Frequency','Monetary']
```

```python
rfm['Recency'] = rfm['Recency'].astype(int)
rfm
```

<h3 style="color:#790699">Finding top 10 customers</h3>


<h4 style="color:#111111">Customers with the lowest recency, highest frequency and monetary amounts considered as top customers.</h4>


```python
#Sorting recency, frequency and monetary ascending, descending and descending respecteviely
rfm.sort_values(['Recency','Frequency', 'Monetary'], ascending=[True,False, False], inplace=True) 
rfm.head(10)
```

```python
quantiles = rfm.quantile(q=[0.25,0.5,0.75])
quantiles
```
